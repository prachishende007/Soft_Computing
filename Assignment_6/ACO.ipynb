```python
import matplotlib.pyplot as plt
import numpy as np
import re

# City coordinates for visualization
# (You can tweak these to visually match the image layout)
coords = np.array([
    [0, 0],   # City 0
    [0, 4],   # City 1
    [4, 2],   # City 2
    [6, 0]    # City 3
])

# Read file
with open("aco_ants_paths.txt") as f:
    lines = f.readlines()

ant_paths = {1: [], 2: [], 3: []}
for line in lines:
    match = re.search(r"Ant (\d+): (.*?)\|", line)
    if match:
        ant_id = int(match.group(1))
        path = list(map(int, match.group(2).strip().split()))
        ant_paths[ant_id] = path

# Plot setup
colors = ['red', 'blue', 'green']
plt.figure(figsize=(6,6))

# Plot city points
plt.scatter(coords[:,0], coords[:,1], color='black', s=100)
for i, (x, y) in enumerate(coords):
    plt.text(x+0.1, y+0.2, f"City {i}", fontsize=9)

# Draw each ants path
for k, color in zip(ant_paths.keys(), colors):
    path = ant_paths[k]
    path_coords = coords[path]
    plt.plot(path_coords[:,0], path_coords[:,1], marker='o', color=color, label=f"Ant {k}")

plt.title("Ant Colony Optimization  Paths of 3 Ants")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print the 3rd ants final path
print("Ant 3 Path:", ant_paths[3])
```
