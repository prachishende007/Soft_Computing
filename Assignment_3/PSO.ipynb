```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/particle_fitness_log.csv')
```

```python
# Plot each particle's fitness
for col in df.columns[1:-1]:
    plt.plot(df["Iteration"], df[col], alpha=0.5, label=col)

# Plot mean fitness
plt.plot(df["Iteration"], df["MeanFitness"], 'k-', linewidth=2, label="Mean Fitness")

plt.title("Particle Fitness Convergence in PSO")
plt.xlabel("Iteration")
plt.ylabel("Fitness Value")
plt.legend()
plt.grid(True)
plt.show()
```
