```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame called df
df = pd.read_csv('/content/gwo_positions.csv')

plt.figure(figsize=(8,6))

# Scatter plot for each role
for role in df['Role'].unique():
    subset = df[df['Role'] == role]
    plt.scatter(subset['X'], subset['Y'], label=role, s=100, alpha=0.7)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('GWO Wolves Positions by Role')
plt.legend()
plt.grid(True)
plt.show()
```
